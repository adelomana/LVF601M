```python
import random
import matplotlib.pyplot as plt
from networkx import nx
import numpy as np 
n = 10  # 10 nodes
m = 20  # 20 edges
# create graph
G = nx.Graph()
# add nodes
### define a list nodes....
node_list = list(range(n))
G.add_nodes_from(node_list)
# add edges
edge_list = []
### define a list of edges. It should be a list of tuples. keep in mind that you should not have 
### the same edge twice and that edge (a, b) is the same than (b, a)
for x in range(m):
    #generate random numbers
    random_edge = list(np.random.choice(n, 2))
    print(random_edge)
    #see if there is already a edge between those nodes or the reverse
    reversed_edge = random_edge[::-1]
    if random_edge and reversed_edge not in edge_list:
        #append
        edge_list.append(random_edge)
    
G.add_edges_from(edge_list)
print(list(G.nodes))
print(list(G.edges))
nx.draw(G)
plt.show()
```
